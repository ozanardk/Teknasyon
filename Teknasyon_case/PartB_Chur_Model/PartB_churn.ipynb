feature engineering

The amount variables have been extracted and expressed as ratios. In the absence of a continuous learning infrastructure, all amount columns can be extracted.

```python
import numpy as np
import pandas as pd

# Feature engineering for churn prediction model.

# Ensure no division by zero
EPSILON = 1e-10

# --- 1. Usage Intensity and Spending Ratios ---
df['calls_outgoing_avg_spending'] = df['calls_outgoing_spendings'] / (df['calls_outgoing_count'] + EPSILON)
df['calls_outgoing_avg_duration'] = df['calls_outgoing_duration'] / (df['calls_outgoing_count'] + EPSILON)
df['sms_outgoing_avg_spending'] = df['sms_outgoing_spendings'] / (df['sms_outgoing_count'] + EPSILON)
df['gprs_avg_spending_per_session'] = df['gprs_spendings'] / (df['gprs_session_count'] + EPSILON)
df['gprs_avg_usage_per_session'] = df['gprs_usage'] / (df['gprs_session_count'] + EPSILON)
# --- 2. Spending vs. Reload Balances ---
df['reload_to_spending_ratio'] = df['reloads_sum'] / (df['user_spendings'] + EPSILON)
df['avg_reload_amount'] = df['reloads_sum'] / (df['reloads_count'] + EPSILON)
df['days_between_reloads'] = df['reloads_inactive_days'] / (df['reloads_count'] + EPSILON)
# --- 3. Call and SMS Usage Proportions ---
df['calls_onnet_ratio'] = df['calls_outgoing_to_onnet_count'] / (df['calls_outgoing_count'] + EPSILON)
df['calls_offnet_ratio'] = df['calls_outgoing_to_offnet_count'] / (df['calls_outgoing_count'] + EPSILON)
df['calls_abroad_ratio'] = df['calls_outgoing_to_abroad_count'] / (df['calls_outgoing_count'] + EPSILON)
df['sms_onnet_ratio'] = df['sms_outgoing_to_onnet_count'] / (df['sms_outgoing_count'] + EPSILON)
df['sms_offnet_ratio'] = df['sms_outgoing_to_offnet_count'] / (df['sms_outgoing_count'] + EPSILON)
df['sms_abroad_ratio'] = df['sms_outgoing_to_abroad_count'] / (df['sms_outgoing_count'] + EPSILON)
df['data_to_call_ratio'] = df['gprs_usage'] / (df['calls_outgoing_duration'] + EPSILON)
# --- 4. Inactivity Ratios ---
df['calls_inactive_ratio'] = df['calls_outgoing_inactive_days'] / (df['user_lifetime'] + EPSILON)
df['sms_inactive_ratio'] = df['sms_outgoing_inactive_days'] / (df['user_lifetime'] + EPSILON)
df['gprs_inactive_ratio'] = df['gprs_inactive_days'] / (df['user_lifetime'] + EPSILON)
# --- 5. Last 100 Transactions Trend Analysis ---
df['last_100_calls_ratio'] = df['last_100_calls_outgoing_duration'] / (df['calls_outgoing_duration'] + EPSILON)
df['last_100_sms_ratio'] = df['last_100_sms_outgoing_count'] / (df['sms_outgoing_count'] + EPSILON)
df['last_100_gprs_ratio'] = df['last_100_gprs_usage'] / (df['gprs_usage'] + EPSILON)

# --- 6. Customer Segmentation ---
df['is_heavy_caller'] = (df['calls_outgoing_count'] > df['calls_outgoing_count'].mean()).astype(int)
df['is_heavy_texter'] = (df['sms_outgoing_count'] > df['sms_outgoing_count'].mean()).astype(int)
df['is_heavy_data_user'] = (df['gprs_usage'] > df['gprs_usage'].mean()).astype(int)
df['is_low_spender'] = (df['user_spendings'] < df['user_spendings'].median()).astype(int)
df['is_new_user'] = (df['user_lifetime'] < 90).astype(int)
df['is_loyal_user'] = (df['user_lifetime'] > 365).astype(int)
df['is_voice_dominant'] = ((df['calls_outgoing_duration'] / (df['gprs_usage'] + EPSILON)) > 1).astype(int)
df['is_data_dominant'] = ((df['gprs_usage'] / (df['calls_outgoing_duration'] + EPSILON)) > 1).astype(int)
df['is_balanced_user'] = (((df['calls_outgoing_duration'] / (df['gprs_usage'] + EPSILON)) - 1).abs() < 0.2).astype(int)    

# Display the first few rows of the new DataFrame with engineered features
df.head()
```

Feature Selection

Business Goal (Gini Score):
To maximize the separation between churners and non-churners, ensuring the model effectively ranks customers by their likelihood of churn while maintaining a strong distinction between high-risk and low-risk customers.

```python
import numpy as np
import pandas as pd
import xgboost as xgb
import shap

from sklearn.feature_selection import RFE, RFECV
from sklearn.metrics import roc_auc_score, make_scorer
from sklearn.model_selection import StratifiedKFold, train_test_split


base_model = xgb.XGBClassifier(random_state=42)
base_model.fit(X_train, y_train)

# ---------------------------------------------------------------
# 1) METHOD 1 - Embedded Feature Importances (Using 'gain').
# ---------------------------------------------------------------
importance_type = "gain"
importances = base_model.get_booster().get_score(importance_type=importance_type)
importance_df = pd.DataFrame(list(importances.items()), columns=["Feature", "Importance"])
importance_df = importance_df.sort_values(by="Importance", ascending=False)

# Selecting features with importance > 0 and limiting to 40
selected_importance_features = importance_df[importance_df["Importance"] > 0]["Feature"].tolist()
selected_importance_features = selected_importance_features[:15]

# ---------------------------------------------------------------
# 2) METHOD 2 - Feature Selection Using SHAP Values
# ---------------------------------------------------------------
explainer = shap.TreeExplainer(base_model)
shap_values = explainer.shap_values(X_train)

shap_mean = np.abs(shap_values).mean(axis=0)
shap_df = pd.DataFrame({
    "Feature": X_train.columns,
    "SHAP_Value": shap_mean
}).sort_values(by="SHAP_Value", ascending=False)

threshold = 0.01
selected_shap_features = shap_df[shap_df["SHAP_Value"] > threshold]["Feature"].tolist()
selected_shap_features = selected_shap_features[:15]

# ---------------------------------------------------------------
# 3) METHOD 3 - RFE Using 'gain' as Feature Importance
# ---------------------------------------------------------------
rfe = RFE(estimator=base_model, n_features_to_select=40, step=1)
rfe.fit(X_train, y_train)
selected_rfe_features = X_train.columns[rfe.support_].tolist()
selected_rfe_features = selected_rfe_features[:15]

# ---------------------------------------------------------------
# 4) METHOD 4 - RFECV Using 'gain' as Selection Criterion
# ---------------------------------------------------------------
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
gini_scorer = make_scorer(roc_auc_score)
rfecv = RFECV(
    estimator=base_model,
    step=1,
    cv=cv,
    scoring=gini_scorer,
    min_features_to_select=10
)
rfecv.fit(X_train, y_train)

rfecv_ranking = pd.DataFrame({
    "Feature": X_train.columns,
    "Rank": rfecv.ranking_
})
rfecv_selected = rfecv_ranking[rfecv_ranking["Rank"] == 1].sort_values("Feature")
if len(rfecv_selected) > 15:
    rfecv_selected = rfecv_selected.head(15)
selected_rfecv_features = rfecv_selected["Feature"].tolist()

# ---------------------------------------------------------------
# 5) Number of Selected Features.
# ---------------------------------------------------------------
print("Number of features selected using embedded importance :", len(selected_importance_features))
print("Number of features selected using SHAP             :", len(selected_shap_features))
print("Number of features selected using RFE               :", len(selected_rfe_features))
print("Number of features selected using RFECV (Gini)         :", len(selected_rfecv_features))

# ---------------------------------------------------------------
# 6) Model performance with the selected feature sets (Train & Test Gini).
# ---------------------------------------------------------------
def train_and_evaluate_features(feature_list, X_train, y_train, X_test, y_test):
    model = xgb.XGBClassifier(random_state=42)
    X_train_sub = X_train[feature_list]
    X_test_sub = X_test[feature_list]
    
    model.fit(X_train_sub, y_train)
    
    y_pred_train = model.predict_proba(X_train_sub)[:, 1]
    gini_train = 2 * roc_auc_score(y_train, y_pred_train) - 1
    
    y_pred_test = model.predict_proba(X_test_sub)[:, 1]
    gini_test = 2 * roc_auc_score(y_test, y_pred_test) - 1
    
    return gini_train, gini_test

# Train and evaluate models with different feature sets
gini_embedded_train, gini_embedded_test = train_and_evaluate_features(
    selected_importance_features, X_train, y_train, X_test, y_test
)
print(f"\nEmbedded Importance (Gain) - Train Gini: {gini_embedded_train:.4f}, Test Gini: {gini_embedded_test:.4f}")

gini_shap_train, gini_shap_test = train_and_evaluate_features(
    selected_shap_features, X_train, y_train, X_test, y_test
)
print(f"SHAP - Train Gini: {gini_shap_train:.4f}, Test Gini: {gini_shap_test:.4f}")

gini_rfe_train, gini_rfe_test = train_and_evaluate_features(
    selected_rfe_features, X_train, y_train, X_test, y_test
)
print(f"RFE - Train Gini: {gini_rfe_train:.4f}, Test Gini: {gini_rfe_test:.4f}")

gini_rfecv_train, gini_rfecv_test = train_and_evaluate_features(
    selected_rfecv_features, X_train, y_train, X_test, y_test
)
print(f"RFECV (Gini) - Train Gini: {gini_rfecv_train:.4f}, Test Gini: {gini_rfecv_test:.4f}")
```

Modeling experiments

```python
import numpy as np
import pandas as pd
import xgboost as xgb

from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
import pickle
import os

all_combined_features = set(selected_importance_features) \
    .union(set(selected_shap_features)) \
    .union(set(selected_rfe_features)) \
    .union(set(selected_rfecv_features))

feature_subsets = {
    "all_combined": list(all_combined_features),
    "embedded_importance": selected_importance_features,
    "shap": selected_shap_features,
    "rfe": selected_rfe_features,
    "rfecv": selected_rfecv_features  
}

# Optimized Parameter lists for better generalization
learning_rate_list = [0.01, 0.05, 0.1]
max_depth_list = [8, 6, 5, 4]
min_child_weight_list = [3, 5, 7]
subsample_list = [0.7, 0.8]
colsample_bytree_list = [0.7, 0.8]
n_estimators_list = [500,450,400,350,300, 200, 150]

# Cross-validation settings (Increased folds for better stability)
N_SPLITS = 5
cv = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

best_gini = 0.0
best_model_file = "best_model.pkl"
best_model_params = None
results = []

for subset_name, subset_cols in feature_subsets.items():
    X_train_sub = X_train[subset_cols]
    X_test_sub = X_test[subset_cols]
    
    for lr in learning_rate_list:
        for md in max_depth_list:
            for mcw in min_child_weight_list:
                for subs in subsample_list:
                    for colsub in colsample_bytree_list:
                        for nest in n_estimators_list:
                            
                            model_cv = xgb.XGBClassifier(
                                random_state=42,
                                eval_metric="auc",
                                learning_rate=lr,
                                max_depth=md,
                                min_child_weight=mcw,
                                subsample=subs,
                                colsample_bytree=colsub,
                                n_estimators=nest,
                                n_jobs=-1
                            )
                            
                            gini_scores = []
                            for train_idx, val_idx in cv.split(X_train_sub, y_train):
                                X_cv_train, X_cv_val = X_train_sub.iloc[train_idx], X_train_sub.iloc[val_idx]
                                y_cv_train, y_cv_val = y_train.iloc[train_idx], y_train.iloc[val_idx]
                                
                                model_cv.fit(X_cv_train, y_cv_train, eval_set=[(X_cv_val, y_cv_val)], verbose=False)
                                y_val_prob = model_cv.predict_proba(X_cv_val)[:, 1]
                                gini_score = 2 * roc_auc_score(y_cv_val, y_val_prob) - 1
                                gini_scores.append(gini_score)
                            
                            cv_mean_gini = np.mean(gini_scores)
                            
                            model_final = xgb.XGBClassifier(
                                random_state=42,
                                eval_metric="auc",
                                learning_rate=lr,
                                max_depth=md,
                                min_child_weight=mcw,
                                subsample=subs,
                                colsample_bytree=colsub,
                                n_estimators=nest,
                                n_jobs=-1
                            )
                            model_final.fit(X_train_sub, y_train, verbose=False)
                            
                            y_train_prob = model_final.predict_proba(X_train_sub)[:, 1]
                            gini_train = 2 * roc_auc_score(y_train, y_train_prob) - 1
                            
                            y_test_prob = model_final.predict_proba(X_test_sub)[:, 1]
                            gini_test = 2 * roc_auc_score(y_test, y_test_prob) - 1
                            
                            if abs(gini_train - gini_test) > 0.05:
                                continue  # Skip models with train-test gap > 0.05
                            
                            result_row = {
                                "subset": subset_name,
                                "learning_rate": lr,
                                "max_depth": md,
                                "min_child_weight": mcw,
                                "subsample": subs,
                                "colsample_bytree": colsub,
                                "n_estimators": nest,
                                "cv_mean_gini": cv_mean_gini,
                                "gini_train": gini_train,
                                "gini_test": gini_test
                            }
                            results.append(result_row)
                            
                            if gini_test > best_gini:
                                best_gini = gini_test
                                if os.path.exists(best_model_file):
                                    os.remove(best_model_file)
                                with open(best_model_file, "wb") as f:
                                    pickle.dump(model_final, f)
                                best_model_params = result_row.copy()
                                
                                print("=" * 60)
                                print(f"New best model saved -> {best_model_file}")
                                print(f"Subset: {subset_name}")
                                print(f"Gini Train Score: {gini_train:.4f}")
                                print(f"Gini Test Score: {gini_test:.4f}")
                                print("=" * 60)

results_df = pd.DataFrame(results)
results_df.to_csv("results_gini.csv", index=False)

print("\nAll experiments complete.")
print(f"Highest Gini Score: {best_gini:.4f}")

if best_model_params:
    print("Best model parameters:")
    for k, v in best_model_params.items():
        print(f" {k}: {v}")
```
